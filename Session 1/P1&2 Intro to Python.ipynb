</span>



In Euclidean three-space, the distance between points $(x_1,y_1,z_1)$ and $(x_2,y_2,z_2)$ is:



<style>img{width:600px;}</style>'



````python
def distance(A,B):
    """
    A and B are lists/tuples containing the x,y,z coordinates of two points in 3-space
    """
    return ( (A[0]-B[0])**2 + (A[1]-B[1])**2 + (A[2]-B[2])**2 )**.5
````

````python
# This code should be saved as "ExampleModules.py"

def squareIt(x):
    return float(x)**2

def sqrtIt(x):
    if x>=0:
        return float(x)**.5
    else :
        return None


class stockBeta():
    def __init__(self, beta):
        self.beta= beta

    # means: __init__ is the constructor (bing bang of the class). When you create a stockBeta object, you pass a beta value, which is stored in the object as self.beta.

    def adjustBeta(self):
        return 0.66*self.beta + 0.33 * 1


    def squareIt(self):
        return self.beta**2

    def __repr__(self):
        return f"stockBeta(beta={self.beta})"

# What Is Stock Beta? Beta measures a stock's volatility compared to the market. Example:
# beta = 1.0: moves like the market, beta > 1.0: more volatile than the market, beta < 1.0: less volatile
# Raw beta (from historical data) can be unreliable long-term.
# Blume Adjustment: It's a weighted average of the raw beta and the market average beta (which is 1.0). 

````

we use **numpy.polyfit(x, y, deg)** for Least squares polynomial fit.

* documentation here: https://numpy.org/doc/stable/reference/generated/numpy.polyfit.html

* The output contains an array for the coefficients ($\beta$ ) from degree 0 to p: $\sum_{i=0:p} \beta_i \times x^{i}  $


````python
# a linear model has degree =1

fitted= np.polyfit(x,y,1)   # fits a polynomial of degree 1 (i.e., a straight line) to the data. [ slope , intercept ]
print(fitted)
````

````python
y_hat =  x * fitted[0] + fitted[1]
````

### Class Excercise 2: Solution

````python

## 
# write a function to do this
# y_bar =
# y_hat =

import matplotlib.pyplot as plt
import numpy as np
err = np.random.random(100) 
x= np.linspace(0,1,100)     
y = 3 + 2 * x + err
fitted= np.polyfit(x,y,1)   
print(fitted)
y_hat =  x * fitted[0] + fitted[1]

plt.plot(x,y,'.')
plt.plot(x,y_hat) 

def r_squared(y , y_hat):
    
    """
    Calculates the R-squared value.
    
    Parameters:
        y (array-like): Actual y values.
        y_hat (array-like): Predicted y values from the model.
    
    Returns:
        float: R-squared value.
    """
    ss_res = np.sum((y - y_hat) ** 2)               # Sum of squared residuals
    ss_tot = np.sum((y - np.mean(y)) ** 2)      # Total sum of squares
    r_squared = 1 - (ss_res / ss_tot)
    return r_squared

r_squared (y, y_hat)
````
