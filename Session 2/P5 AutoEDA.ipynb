```python
df = pd.concat([
        pd.read_csv('Input/Titanic/test.csv'),
        pd.read_csv('Input/Titanic/train.csv')
    ], axis=0)\
    .drop('Survived', axis=1)\
    .merge(
        pd.read_csv('Input/Titanic/gender_submission.csv'),
        on='PassengerId'
    )

df.head()
```

```python
import re
df['Title']=df.Name.apply(lambda x: re.findall(', \w+.',x)[0][1:])
```

```python
import pandas as pd
print("Pandas version:", pd.__version__)
import jinja2
print("Jinja2 version:", jinja2.__version__)    
import autoviz
print("AutoViz version:", autoviz.__version__)
```

```python
from autoviz.AutoViz_Class import AutoViz_Class

AV = AutoViz_Class()
dftc = AV.AutoViz(
    filename='',
    sep='' ,
    depVar='Survived',
    dfte=df,
    header=0,
    verbose=1,
    lowess=False,
    chart_format='png',
    max_rows_analyzed=300000,
    max_cols_analyzed=30
)
```

```python
print('AutoViz finished!!')
finish_time = dt.datetime.now()
print("Finished at ", finish_time)
elapsed = finish_time - start_time
print("Elapsed time: ", elapsed)
```
